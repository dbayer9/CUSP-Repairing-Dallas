```python
with open('./file/validList.json', 'r', encoding='gbk') as fp:
    dataJson = json.load(fp)
    realUrlList = []
    for item in dataJson:
        searchVal = item['name'] + ',' + item['address']
        urll =  'https://www.google.com/maps/search/' + searchVal + '/@' + str(item['lon']) + ',' + str(item['lat']) + ',11z/data=!3m1!4b1'
        urll = urll.replace(' ', '+')
        print(urll)
        driver.get(urll)
        time.sleep(3)
        try:
            xp = '//a[@class="hfpxzc"]'
            item = driver.find_element_by_xpath(xp)
            realUrl = item.get_attribute('href')
            realUrlList.append(realUrl)
        except:
             realUrlList.append(urll)
        time.sleep(3)
```

```python
for item in range(0, len(dataJson)):
    dataJson[item]['realUrl'] = realUrlList[item]
```
