```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy.stats import zscore
import math
```

```python
scores = pd.read_csv('/content/drive/MyDrive/03 Courses/02 Spring/2022 CUSP - Capstone project/05 Pipeline/01 MSA/01 Data/01_MSA_parameters.csv')
scores = scores[['GEO_ID','kitch_perc','plumbing_perc','Median_Age','Median_Income']]
scores
```

```python
MSA_CT_full = pd.read_csv('/content/drive/MyDrive/03 Courses/02 Spring/2022 CUSP - Capstone project/05 Pipeline/01 MSA/01 Data/02_MSA_CTs_full.csv')
MSA_CT_full = MSA_CT_full[['id','Estimate!!HOUSING OCCUPANCY!!Total housing units']]
MSA_CT_full.rename({'Estimate!!HOUSING OCCUPANCY!!Total housing units': 'Total_housing_units'}, axis='columns', inplace=True)
MSA_CT_full
```

```python
scores_mrg = pd.merge(scores, MSA_CT_full, left_on='GEO_ID', right_on='id')
scores_mrg = scores_mrg[['GEO_ID','kitch_perc','plumbing_perc','Median_Age','Median_Income','Total_housing_units']]
scores_mrg
```

```python
scores_expand = scores_mrg.loc[scores_mrg.index.repeat(scores_mrg['Total_housing_units'])]
scores_expand
```

```python
scores_expand['kitch_zscore']     =  (zscore(scores_expand['kitch_perc']))
scores_expand['plumbing_zscore']  =  (zscore(scores_expand['plumbing_perc']))
scores_expand['age_zscore']       =  (zscore(scores_expand['Median_Age']))
scores_expand['income_zscore']    =  (zscore(scores_expand['Median_Income']))
scores_expand
```

```python
scores_grp = scores_expand.groupby(['GEO_ID']).agg({'kitch_perc':'first',
                                                    'plumbing_perc':'first',
                                                    'Median_Age':'first',
                                                    'Median_Income':'first',
                                                    'kitch_zscore':'first',
                                                    'plumbing_zscore':'first',
                                                    'age_zscore':'first',
                                                    'income_zscore':'first',
                                                    'Total_housing_units':'first',})

scores_grp['kitch_zscore'] = scores_grp['kitch_zscore'] * (-1)
scores_grp['plumbing_zscore'] = scores_grp['plumbing_zscore'] * (-1)
scores_grp['age_zscore'] = scores_grp['age_zscore'] * (-1)

scores_grp.reset_index(level=0, inplace=True)
scores_grp
```
